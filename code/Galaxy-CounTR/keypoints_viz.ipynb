```python
import cv2
import matplotlib.pyplot as plt
from PIL import Image
from skimage.measure import label, regionprops
import json
import os
import matplotlib

PATH_TO_ANNOT = 'path/to/annotations.json'
PATH_TO_IMG_FOLDER = 'path/to/images/'

with open(PATH_TO_ANNOT) as f:
    annotations = json.load(f)

images = os.listdir(PATH_TO_IMG_FOLDER)
```

```python
def vis_keypoints(image_id):
    image = cv2.imread(PATH_TO_IMG_FOLDER + image_id)

    for (x, y) in annotations[image_id]['points']:
        cv2.circle(image, (int(x), int(y)), 2, (0, 255, 0), -1)

    for box in annotations[image_id]['box_examples_coordinates']:
        cv2.rectangle(image, box[1], box[3], (255, 0, 0), 2)

    plt.figure(figsize=(16, 16))
    plt.axis('off')
    plt.imsave('./test.png', image)
```

```python
image_id = '000001.png'
vis_keypoints(image_id)
```
